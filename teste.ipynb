```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.getOrCreate()

print(spark)


df = spark.read.csv("./bmw_global_sales_2018_2025.csv", header=True, inferSchema=True)

df2 = df.withColumnsRenamed(
    {
        "Year": "Ano",
        "Month": "Mes",
        "Region": "Regiao",
        "Model": "Modelo",
        "Units_Sold": "Unidades_vendidas",
        "Avg_Price_EUR": "Preco_medio_EUR",
        "Revenue_EUR": "Rev_EUR",
        "BEV_Share": "BEV_Porcentagem",
        "Premium_Share": "Porcentagem_Premiun",
        "GDP_Growth": "Crescimento_Growth",
        "Fuel_Price_Index": "Preco_Fixo_Combustivel",
    }
)
df_filter = df2.filter(df2["Ano"] != 2018)

df_filter.show(n=100)

# df.printSchema()
```
